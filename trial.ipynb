```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.decomposition import PCA
```

```python
path = '/content/drive/MyDrive/train.csv'
data_frame = pd.read_csv(path)

print(data_frame.columns)
```

```python
data = data_frame.drop(columns=['DATE (MM/DD)', 'MST', 'Global CMP22 (vent/cor) [W/m^2]',
       'Direct sNIP [W/m^2]', 'Azimuth Angle [degrees]',
       'Tower Dry Bulb Temp [deg C]', 'Tower Wet Bulb Temp [deg C]',
       'Tower Dew Point Temp [deg C]', 'Avg Wind Direction @ 6ft [deg from N]',
       'Station Pressure [mBar]','Snow Depth [cm]','Albedo (CMP11)'])
data1 = data.loc[data['Total Cloud Cover [%]']>-1]
data1 = data1.loc[data['Total Cloud Cover [%]']<101]
data2 = data1.iloc[1:,]
data1 = data1.iloc[:data1.shape[0]-1,]
# data1
# dataTemp = data_frame.drop(columns=['DATE (MM/DD)', 'MST', 'Global CMP22 (vent/cor) [W/m^2]',
#        'Direct sNIP [W/m^2]', 'Azimuth Angle [degrees]', 'Tower RH [%]','Peak Wind Speed @ 6ft [m/s]', 'Avg Wind Direction @ 6ft [deg from N]',
#        'Station Pressure [mBar]','Snow Depth [cm]','Albedo (CMP11)','Precipitation (Accumulated) [mm]','Moisture'])
# dataTemp = dataTemp.loc[dataTemp['Total Cloud Cover [%]']>-1]
# dataTemp = dataTemp.loc[dataTemp['Total Cloud Cover [%]']<101]
# dataTemp = dataTemp.drop(columns=['Total Cloud Cover [%]'])
# pca1 = PCA(n_components=1)
# dataTemp = pca1.fit_transform(X=dataTemp)
```

```python
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
X_train, y_train = data1,data2
rf = RandomForestRegressor(n_estimators = 500, random_state = 42)
rf.fit(X_train, y_train)
```

```python
def test_res(min,f_1):
  f = f_1
  for i in range(min):
    f1 = rf.predict(np.array(f).reshape(1,-1))
    f = f1
  return f
```

```python
result = []
for j in range(1,301):
  path = '/content/drive/MyDrive/test/'+str(j)+'/weather_data.csv'
  test_data = pd.read_csv(path)
  test_data['Time [Mins]'] = test_data['Time [Mins]'] + 7*60+40
  test_data = test_data.drop(columns=['Global CMP22 (vent/cor) [W/m^2]','Direct sNIP [W/m^2]', 'Azimuth Angle [degrees]','Precipitation (Accumulated) [mm]','Snow Depth [cm]', 'Moisture', 'Albedo (CMP11)'])
  test = test_data.drop('Total Cloud Cover [%]',axis=1)
  c = test_data.drop('Time [Mins]',axis=1)
  tim = int(test['Time [Mins]'].iloc[test.shape[0]-1])
  res_30 = test_res(tim,30,test,c)
  res_60 = test_res(tim,60,test,c)
  res_90 = test_res(tim,90,test,c)
  res_120 = test_res(tim,120,test,c)
  print(res_30)
  print(res_60)
  print(res_90)
  print(res_120)
  tem_res = [res_30[:,4],res_60[:,4],res_90[:,4],res_120[:,4]]
  result.append(tem_res)
```
